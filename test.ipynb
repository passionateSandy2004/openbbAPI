```python
# Historical Data Fetching using KiteConnect
from kiteconnect import KiteConnect
from datetime import datetime, timedelta
import pandas as pd

# Initialize KiteConnect (different from KiteTicker)
# API Key and Access Token (same as used for live data)
api_key = "6ej0i7fspon9w0yv"
access_token = "e8QB3Kbm4dVWFENBFAVSRH1aVmGHtCe4"

kite = KiteConnect(api_key=api_key)
kite.set_access_token(access_token)

# Define instrument token (same as used in live data)
# 738561 - example instrument token from your live data
instrument_token = 738561

# Define date range for historical data
# Get data for last 30 days
to_date = datetime.now()
from_date = to_date - timedelta(days=30)

# Available intervals: minute, 3minute, 5minute, 10minute, 15minute, 30minute, 
#                     60minute, day, 3day, week, month
interval = "day"  # Options: minute, 3minute, 5minute, 15minute, 30minute, 60minute, day

# Fetch historical data
# Parameters:
# - instrument_token: The instrument token (738561 in your case)
# - from_date: Start date (datetime object)
# - to_date: End date (datetime object)
# - interval: Time interval (minute, day, etc.)
# - continuous: For futures/options, set to 1 for continuous contract, 0 for current contract
# - oi: Set to 1 to include Open Interest data (for F&O instruments)

try:
    historical_data = kite.historical_data(
        instrument_token=instrument_token,
        from_date=from_date,
        to_date=to_date,
        interval=interval,
        continuous=0,  # 0 for current contract, 1 for continuous
        oi=0  # 1 to include Open Interest, 0 to exclude
    )
    
    # Convert to DataFrame for easier analysis
    df = pd.DataFrame(historical_data)
    
    # Convert date column to datetime
    df['date'] = pd.to_datetime(df['date'])
    
    print(f"Fetched {len(df)} records")
    print("\nFirst few rows:")
    print(df.head())
    print("\nLast few rows:")
    print(df.tail())
    print("\nDataFrame Info:")
    print(df.info())
    
except Exception as e:
    print(f"Error fetching historical data: {e}")
```

```python
# Example: Fetching different time intervals

# Example 1: Get 1-minute data for last 5 days
def get_minute_data(instrument_token, days=5):
    to_date = datetime.now()
    from_date = to_date - timedelta(days=days)
    
    data = kite.historical_data(
        instrument_token=instrument_token,
        from_date=from_date,
        to_date=to_date,
        interval="minute",
        continuous=0,
        oi=0
    )
    return pd.DataFrame(data)

# Example 2: Get 5-minute data for last 10 days
def get_5minute_data(instrument_token, days=10):
    to_date = datetime.now()
    from_date = to_date - timedelta(days=days)
    
    data = kite.historical_data(
        instrument_token=instrument_token,
        from_date=from_date,
        to_date=to_date,
        interval="5minute",
        continuous=0,
        oi=0
    )
    return pd.DataFrame(data)

# Example 3: Get daily data for last 1 year
def get_daily_data(instrument_token, days=365):
    to_date = datetime.now()
    from_date = to_date - timedelta(days=days)
    
    data = kite.historical_data(
        instrument_token=instrument_token,
        from_date=from_date,
        to_date=to_date,
        interval="day",
        continuous=0,
        oi=0
    )
    return pd.DataFrame(data)

# Example 4: Get data with Open Interest (for F&O instruments)
def get_data_with_oi(instrument_token, days=30):
    to_date = datetime.now()
    from_date = to_date - timedelta(days=days)
    
    data = kite.historical_data(
        instrument_token=instrument_token,
        from_date=from_date,
        to_date=to_date,
        interval="day",
        continuous=0,
        oi=1  # Include Open Interest
    )
    return pd.DataFrame(data)

# Uncomment to test:
# df_minute = get_minute_data(738561, days=5)
# print("1-minute data:", df_minute.head())

# df_daily = get_daily_data(738561, days=365)
# print("Daily data:", df_daily.head())
```
